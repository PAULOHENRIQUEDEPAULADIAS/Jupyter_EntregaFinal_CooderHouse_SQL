## Resolução

### Introdução

Este projeto visa simular o banco de dados de um serviço de Streaming. Para fins deste projeto, não foram considerados variáveis como: Usuário, Atores, Diretores, Ficheiros, etc. Foram considerados apenas o banco de dados para os títulos.
O serviço possibilita a visualização do banco, a ordenação por categoria, como ano de lançamento ou popularidade por exemplo, a remoção e adição de títulos ao banco de dados garantindo a segurança de se saber quais títulos foram adicionados e removidos. Também foi pensado na facilitação do serviço de busca, possibilitando a busca do título por ID ou a overview por título.

### Definições

* Descrição dos bancos de dados:
    * top_rated_movies: Banco de dados que armazena os títulos dos fílmes, possui os seguintes campos:
        * id, title, overview, release_date, popularity, vote_average, vote_count.
    * palavras_genero: Banco de dados transacional com armazenamento de informações sobre gênero de filmes, relacionando-os com palavras específicas, eg: elf = fantasy, passion = romance, etc.
    * filmes_excluidos: Banco de dados destinado ao armazenamento de títulos excluidos do banco de dados principal.
    * filmes_adicionados: Banco de dados destinado ao armazenamento de títulos adicionados ao banco de dados principal.

* Descrição das views:
    * consulta_de_genero: View destinada a relacionar o banco de dados principal com o banco de dados palavras_genero e nos devolver o gênero sugerido do filme.
    * consulta_por_ano_de_lancamento: View destinada a visualizar de forma ordenada o banco de dados, ordenado por ano de lançamento.
    * consulta_por_contagem_de_votos: View destinada a visualizar de forma ordenada o banco de dados, ordenado por contagem de votos.
    * consulta_por_media_de_votos: View destinada a visualizar de forma ordenada o banco de dados, ordenado por média de votos.
    * consulta_por_popularidade: View destinada a visualizar de forma ordenada o banco de dados, ordenado por popularidade do filme.

* Descrição dos Stored Procedures:
    * add_movie: Stored Procedure destinada a adicionar informações de novos filmes no banco de dados top_rated_movies.
    * delete_movie: Stored Procedure destinada a deletar um filme especifico do banco de dados top_rated_movies.

* Descrição das Functions:
    * retorno_overview_por_titulo: Function destinada a retornar a overview de um título específico.
    * retorno_titulo_por_id: Function destinada a retornar o título do filme.

* Descrição das Triggers:
    * antes_de_adicionar: Trigger destinada a realizar uma cópia dos dados que estão sendo inseridos pelo stored procedure, add_movie, e os armazena dentro do banco de dados filmes_adicionados.
    * depois_de_adicionar: Trigger destinada a realizar uma cópia dos dados que estão sendo deletados pelo stored procedure, delete_movie, e os armazena dentro do banco de dados filmes_excluidos.
